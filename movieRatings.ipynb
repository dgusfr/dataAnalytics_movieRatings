# Ratings

O presente projeto utiliza como base um conjunto de dados robusto, composto por **100.836 registros** de avaliações de filmes realizadas por usuários em uma plataforma de recomendação. Este dataset, amplamente reconhecido em estudos de sistemas de recomendação e análise de comportamento do usuário, é estruturado em quatro colunas principais:



*   **userId**: Identificador exclusivo para cada usuário que realizou uma avaliação, permitindo o acompanhamento de padrões individuais de consumo.
*   **movieId**: Identificador único para cada filme presente na base de dados, possibilitando análises detalhadas sobre a popularidade e o desempenho de diferentes títulos.
*   **rating**: Nota atribuída pelos usuários aos filmes, variando de 0,5 a 5, e representando uma métrica direta de preferência ou satisfação.
*   **timestamp**: Registro temporal em formato Unix, que indica o momento exato em que a avaliação foi realizada, permitindo a análise de tendências temporais.


O objetivo principal do projeto é explorar este conjunto de dados para realizar análises detalhadas, identificar padrões de comportamento e preferências dos usuários, e construir modelos de recomendação baseados em similaridade e tendências observadas.

Além disso, a granularidade e a escala do **dataset** permitem abordar questões como sazonalidade nas avaliações, identificação de usuários com comportamentos extremos e a validação de **hipóteses** sobre correlações entre diferentes variáveis.

Este relatório detalha o processo de análise, incluindo a **preparação**, **limpeza** e **visualização** dos dados, bem como os insights obtidos a partir dos métodos analíticos aplicados.

```python
medias_por_filme.describe()
```

```python
sns.displot(medias_por_filme, kde=True)
plt.title("Histograma das médias dos filmes")
```

O histograma das médias dos filmes exibe a **distribuição das avaliações médias **atribuídas pelos usuários aos filmes. Com a linha de densidade sobreposta (kde=True), é possível observar a forma da distribuição de maneira mais clara.

Em geral, gráficos desse tipo revelam padrões como a presença de picos em torno de certas médias (como **3 ou 4**, em escalas de 1 a 5), indicando que os filmes tendem a ser avaliados de forma regular, com menos frequência em extremos como notas muito baixas ou muito altas.

Isso sugere que a maioria das avaliações é **moderada**, mas pode haver exceções que destacam filmes particularmente amados ou rejeitados pelos espectadores. Analisar esses padrões pode ajudar a identificar tendências e outliers.

```python
total_geral = len(tmdb)
total_de_ingles = tmdb['original_language'].value_counts()['en']
total_do_resto = total_geral - total_de_ingles

print(total_geral, total_de_ingles, total_do_resto)
```

```python
dados = {
  "lingua" : ["ingles", "outros"],
  "total" : [total_de_ingles, total_do_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(data=dados, x="lingua", y="total")
```

```python
dados.plot(kind="pie", labels=dados["lingua"], y="total")
```
